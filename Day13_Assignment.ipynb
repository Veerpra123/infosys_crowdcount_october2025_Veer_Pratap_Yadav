```python
# Day 13  Matplotlib (Summary + Small Demos)

"""
Goals Today:
- Understand what Matplotlib is and when to use it.
- Learn the basic plotting workflow:
  1) import matplotlib.pyplot as plt
  2) Prepare data (x, y)
  3) Create a figure -> plt.figure()
  4) Plot -> plt.plot()/plt.bar()/plt.scatter()
  5) Label -> plt.title(), plt.xlabel(), plt.ylabel()
  6) Show -> plt.show()

Tips:
- One chart per figure (avoid clutter).
- Keep axes titles and units clear.
- Use legends only when needed.
"""

import matplotlib.pyplot as plt

# ---- Demo 1: Line Plot ----
x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 6, 7]
plt.figure()
plt.plot(x, y)
plt.title("Line Plot  Sales over Days")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()

# ---- Demo 2: Bar Chart ----
categories = ["A", "B", "C", "D"]
values = [10, 6, 8, 12]
plt.figure()
plt.bar(categories, values)
plt.title("Bar Chart  Category Counts")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# ---- Demo 3: Scatter Plot ----
x = [10, 20, 30, 40, 50]
y = [12, 22, 19, 35, 42]
plt.figure()
plt.scatter(x, y)
plt.title("Scatter  Height vs Score (Example)")
plt.xlabel("Height")
plt.ylabel("Score")
plt.show()
```
