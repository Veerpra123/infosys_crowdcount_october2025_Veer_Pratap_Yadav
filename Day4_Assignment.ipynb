```python
# Day 4 Assignment

# Task 1: Calculator with if-elif-else
a = 10
b = 5
op = "+"

if op == "+":
    print("Result:", a + b)
elif op == "-":
    print("Result:", a - b)
elif op == "*":
    print("Result:", a * b)
elif op == "/":
    print("Result:", a / b)
elif op == "%":
    print("Result:", a % b)
elif op == "//":
    print("Result:", a // b)
elif op == "**":
    print("Result:", a ** b)
else:
    print("Invalid operator")

# Task 2: for loop with tuple, dictionary, set
print("\nTuple Loop:")
for x in (1, 2, 3):
    print(x)

print("\nDictionary Loop:")
my_dict = {"a": 1, "b": 2}
for k, v in my_dict.items():
    print(k, ":", v)

print("\nSet Loop:")
for x in {10, 20, 30}:
    print(x)

# Task 3: Calculator with match-case (Python 3.10+)
a, b = 8, 4
op = "*"

match op:
    case "+":
        print("Result:", a + b)
    case "-":
        print("Result:", a - b)
    case "*":
        print("Result:", a * b)
    case "/":
        print("Result:", a / b)
    case "%":
        print("Result:", a % b)
    case "//":
        print("Result:", a // b)
    case "**":
        print("Result:", a ** b)
    case _:
        print("Invalid operator")

# Task 4: File concepts
with open("example.txt", "w") as f:
    f.write("Hello, file handling in Python!")

with open("example.txt", "r") as f:
    print("File Content:", f.read())
```
