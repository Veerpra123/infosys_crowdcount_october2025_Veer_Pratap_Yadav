```python
# Task 1: Calculator using OOP concepts

class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def add(self):
        return self.a + self.b
    
    def subtract(self):
        return self.a - self.b
    
    def multiply(self):
        return self.a * self.b
    
    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Division by zero not allowed"

# Example usage
calc = Calculator(10, 5)
print("Addition:", calc.add())
print("Subtraction:", calc.subtract())
print("Multiplication:", calc.multiply())
print("Division:", calc.divide())
```

```python
# Task 2: Inheritance Examples

# 1. Multiple Inheritance
class A:
    def method_A(self):
        print("Method A")

class B:
    def method_B(self):
        print("Method B")

class C(A, B):
    def method_C(self):
        print("Method C")

obj = C()
obj.method_A()
obj.method_B()
obj.method_C()

print("\n")

# 2. Multilevel Inheritance
class Parent:
    def parent_method(self):
        print("Parent method")

class Child(Parent):
    def child_method(self):
        print("Child method")

class GrandChild(Child):
    def grandchild_method(self):
        print("GrandChild method")

obj2 = GrandChild()
obj2.parent_method()
obj2.child_method()
obj2.grandchild_method()

print("\n")

# 3. Hierarchical Inheritance
class Parent1:
    def parent1_method(self):
        print("Parent1 method")

class Child1(Parent1):
    def child1_method(self):
        print("Child1 method")

class Child2(Parent1):
    def child2_method(self):
        print("Child2 method")

c1 = Child1()
c2 = Child2()
c1.parent1_method()
c1.child1_method()
c2.parent1_method()
c2.child2_method()

print("\n")

# 4. Hybrid Inheritance (Multiple + Multilevel)
class Base:
    def base_method(self):
        print("Base method")

class D(Base):
    def method_D(self):
        print("Method D")

class E(Base):
    def method_E(self):
        print("Method E")

class F(D, E):
    def method_F(self):
        print("Method F")

obj3 = F()
obj3.base_method()
obj3.method_D()
obj3.method_E()
obj3.method_F()
```

```python
# Task 3: Overriding & Overloading

# Method Overriding
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        print("Hello from Child")  # overrides parent method

obj = Child()
obj.greet()

print("\n")

# Method Overloading (Python doesn't support traditional overloading)
# But we can achieve it using default arguments or *args

class Math:
    def add(self, a, b=None):
        if b is not None:
            return a + b
        else:
            return a + 10  # default addition

math_obj = Math()
print(math_obj.add(5, 7))  # 12
print(math_obj.add(5))     # 15 (default)
```
